```python
class CartItem:
  def __init__(self, name, quantity, price):
    self._name = name
    self._price = price
    self._quantity = quantity
  def update_quantity(self, new_quantity):
    self._quantity = new_quantity
  def total_cost(self):
    return self._price * self._quantity
  def print_item(self):
    print(f'name: {self._name}; price: {self._price}; quantity: {self._quantity}')



y = CartItem('aaa', 10, 11)
y.print_item()
print(y.total_cost())
y.update_quantity(11)
y.print_item()
print(y.total_cost())
```

```python
class Automobile:
  class Engine():
    def __init__(self, power):
      self.power = power
    def info(self):
      print(self.power)
  def __init__(self, power):
    self.eng = self.Engine(power)
    self.status = False
  def start(self):
    self.status = True
  def stop(self):
    self.status = False
  def inform(self):
    print(' : ')
    self.eng.info()
    if self.status:
      print(" ")
    else:
      print("  ")
a = Automobile(100)
a.start()
a.inform()
a.stop()
a.inform()
```

```python
from abc import ABC, abstractmethod
class TwoDShape(ABC):
  @abstractmethod
  def area(slef):
    pass
  @abstractmethod
  def perimetr(self):
    pass
class Square(TwoDShape):
  def __init__(self, val):
    self.sideLength = val
  def area(self):
    return self.sideLength ** 2
  def perimetr(self):
    return self.sideLength * 4


a = Square(5)
print(a.area(), a.perimetr())
```

```python
from abc import ABC, abstractmethod
class ICatrItem(ABC):
  @abstractmethod
  def updateQuantity(self):
    pass
  @abstractmethod
  def getTotalPrice(self):
    pass


class CatrItem(ICatrItem):
  def __init__(self, productName, quantity, price):
    self.productName = productName
    self.quantity = quantity
    self.price = price
  def updateQuantity(self, newQuantity):
    self.quantity = newQuantity
  def getTotalPrice(self):
    return self.price * self.quantity
  def disp(self):
    print(f'name: {self.productName}; quantity: {self.quantity}; price: {self.price}')

a = CatrItem('aaa', 10, 5)
print(a.getTotalPrice())
a.updateQuantity(7)
print(a.getTotalPrice())
a.disp()
```

```python
class Gadget():
  def __init__(self, spec):
    self.spec = spec
  def getSpecification(self):
    return self.spec

class Smartphone(Gadget):
  def getSpecification(self):
    return 'Smartphone specification: ' + self.spec

class Laptop(Gadget):
  def getSpecification(self):
    return 'Laptop specification: ' + self.spec

a = Gadget('qqq')
b = Smartphone('www')
c = Laptop('eee')
print(a.getSpecification())
print(b.getSpecification())
print(c.getSpecification())
```

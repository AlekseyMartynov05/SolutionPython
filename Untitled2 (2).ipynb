```python
#1
string = input()
if len(string) % 2 != 0:
  s1 = string[:len(string) // 2]
  s2 = string[len(string) // 2 + 1:len(string)]
  string = s1 + s2
else:
  s1 = string[:len(string) // 2 - 1]
  s2 = string[len(string) // 2 + 1:len(string)]
  string = s1 + s2
print(string)
```

```python
#2
string = input()
res = {}
res['*'] = string.count('*')
res[';'] = string.count(';')
res[':'] = string.count(':')
print(res)
```

```python
#3
text = input()
abc = 'абвгдеёжзийклмнопрстуфхцчшщьыъэюя'
res = {}
for i in abc:
  res[i] = text.count(i)
ans = ''
for i in res.keys():
  if res[i] != 0:
    ans += f'{i} - {res[i]}'
    if (res[i] < 10 and 2 <= res[i] <= 4) or (res[i] > 20 and 2 <= res[i] % 10 <= 4):
      ans += ' раза, '
    else:
      ans += ' раз, '
print(ans)
```

```python
#1
from random import randint
path = input()
n = int(input())
if path == '':
    path = 'default.txt'
with open(path, 'w') as file:
    nums = []
    d = n
    while d > 0:
        nums.append(randint(-n, n))
        d -= 1
    for i in nums:
        file.write(str(i) + ' ')
with open(path, 'r') as f:
    line = f.readline()
    line = line.split()
    for i in range(len(line)):
        line[i] = int(line[i])
    c = []
    k = 0
    for i in line:
        if -1 * i in line:
            if not i in c:
                c.append(i)
                c.append(-1 * i)
                k += 1
    print(k)
    print(line)
```

```python
#3
path = 'Bibl.txt'
try:
    with open(path, 'r') as file:
        lines = file.readlines()
        count = 0
        for line in lines:
            line = line.split()
            i = 2
            while i < len(line) and not line[i].isdigit():
                i += 1
            if int(line[i]) >= 1960:
                count += 1
        print(count)
except FileNotFoundError:
    print('Файл Bibl.txt не найден.')
```

```python
#3
import re
try:
    with open('input.txt', 'r', encoding='utf-8') as input_file:
        text = input_file.read()
        sentences = re.split('([!.])', text)
        result = []
        for i in sentences:
            i = i.strip()
            words = re.findall(r'(\w+|[^\w\s]+)', i)
            unique_words = []
            for word in words:
               if word not in unique_words and word.capitalize() not in unique_words:
                   unique_words.append(word)
            result.append(unique_words)
        out = ''
        out += result[0][0]
        for i in range(1, len(result)):
            if not result[i] == '.' and not result[i] == '!':
                for j in result[i]:
                    if 'А' <= j <= 'Я' or 'а' <= j <= 'я':
                        out += ' ' + j
                    else:
                        out += j
            else:
                out += result[i][0]
        with open('out.txt', 'w') as output_file:
            output_file.write(out)

except FileNotFoundError:
    print('Файл не найден.')
```

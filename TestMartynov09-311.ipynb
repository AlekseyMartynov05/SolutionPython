```python
from random import shuffle
import datetime
class Question:
    def __init__(self, question, answers, correct_answer):
        self.question = question
        self.answers = answers
        self.correct_answer = correct_answer
    def show(self, number = 0, count = 0):
        if number == 0:
            print(self.question)
        else:
            if count == 0:
                print(f'{number}. {self.question}')
            else:
                print(f'{number}/{count}. {self.question}')
        for i in range(len(self.answers)):
            print(f'{i + 1}. {self.answers[i]}')
    def ask(self, number = 0, count = 0):
        self.show(number, count)
        ans = input()
        flag = True
        while flag:
            if ans == '':
                print('Введите ответ.')
                ans = input()
            elif len(ans) != 1 or (len(ans) == 1 and not ans in '12345'):
                print('Введите число от 1 до 5.')
                ans = input()
            else:
                flag = False
        ans = int(ans)
        if self.answers[ans - 1] in self.correct_answer:
            print('Верно!')
            return True
        else:
            print('Неверно!')
            return False
    def mix(self):
        shuffle(self.answers)



def correct_answer(ans):
    res = ''
    for i in ans:
        if i in '12345':
            res += i
    return res



def form_an_array(text):
    i = 0
    result = []
    while i < len(text) - 6:
        q = text[i][3:]
        if q == '' or q == '\n':
            return False
        ans = []
        for j in range(i + 1, i + 6):
            ans.append(text[j][3:])
            if text[j][3:] == '?' or text[j][3:] == '' or text[j][3:] == '\n':
                return False
        if not text[i + 6][:6] == 'Ответ:':
            return False
        cor_ans_ind = correct_answer(text[i + 6])
        if cor_ans_ind == '':
            return False
        cor_ans = ''
        if len(cor_ans_ind) == 1:
            cor_ans += ans[int(cor_ans_ind) - 1]
        else:
            for ind in cor_ans_ind:
                cor_ans += ans[int(ind) - 1] + ' '
        w = Question(q, ans, cor_ans)
        result.append(w)
        i += 10
    return result



def main():
    path = 'questions.txt'
    if path[-3:] == 'txt':
        try:
            with open(path, 'r', encoding='utf-8') as input_file:
                text = input_file.read()
                text = text.split('\n')
                q_qr = form_an_array(text)
                if not q_qr == False:
                    shuffle(q_qr)
                    correct = 0
                    start = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                    for i in range(len(q_qr)):
                        q_qr[i].mix()
                        if q_qr[i].ask(i + 1, len(q_qr)):
                            correct += 1
                    ending = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                    print('Тестирование завершено!')
                    print(f'Общее количество вопросов: {len(q_qr)}')
                    print(f'Количество правильных ответов: {correct}')
                    print(f'Процент правильных ответов: {correct / len(q_qr) * 100}%')
                    with open('results.txt', 'w') as output_file:
                        output_file.write(start + '\n')
                        output_file.write(ending + '\n')
                        out = 'Общее количество вопросов: ' + str(len(q_qr)) + '\n'
                        out += 'Количество правильных ответов: ' + str(correct) + '\n'
                        out += 'Процент правильных ответов: ' + str(correct / len(q_qr) * 100) + '%'
                        output_file.write(out)
                else:
                    print('Неверный формат данных.')
        except FileNotFoundError:
            print(f'Файл {path} не найден.')
    else:
        print('Неверное расширение файла.')




if __name__ == '__main__':
    main()
```

```python
#2
from math import sin, cos
n = int(input())
arg = 1
res = 1
num = cos(1)
den = sin(1)
while arg <= n:
    res = res * (num / den)
    arg += 1
    num += cos(arg)
    den += sin(arg)
print(res)
```

```python
#3
eps = 0.001
n = 1
el = 1/((3*n-2)*(3*n+1))
s = el
while el > eps:
    n += 1
    el = 1/((3*n-2)*(3*n+1))
    s += el
print(s)
```

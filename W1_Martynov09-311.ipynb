```python
#3
from math import sqrt, acos
def distance(x1, y1, x2, y2):
    return sqrt((x1-x2)**2+(y1-y2)**2)
x1 = float(input())
y1 = float(input())
x2 = float(input())
y2 = float(input())
x3 = float(input())
y3 = float(input())
ab = distance(x1, y1, x2, y2)
ac = distance(x1, y1, x3, y3)
bc = distance(x2, y2, x3, y3)
result = ab == ac + bc or ac == ab + bc or bc == ab + ac
if result:
    print("Точки лежат на одной прямой")
else:
    abc = (ab**2 + bc**2 - ac**2) / (2*ab*bc)
    print(f'Точки не лежат на одной прямой; угол ABC = {acos(abc)}')
```

```python
#1
def sum(num):
    result = 0
    while num != 0:
        result = result + num % 10
        num = num // 10
    return result
def generate(k, n):
    result = []
    for i in range(1, n + 1):
        if sum(i) == k:
            result.append(i)
    return result
def main():
    k = int(input())
    n = int(input())
    print(generate(k, n))
main()
```

```python
#2
def lcm(a, b, c):
    i = max(a, b, c)
    while True:
        if i % a == 0 and i % b == 0 and i % c == 0:
            break
        i = i + 1
    return i
a = int(input())
b = int(input())
c = int(input())
print(lcm(a, b, c))
```

```python
#3
def f(x):
    return x*x - 2
def Root(func, a, b, eps):
    c = (a + b) / 2
    if abs(b - a) < eps:
        return c
    if func(c) == 0:
        return c
    if func(a) * func(c) < 0:
        return Root(func, a, c, eps)
    else:
        return Root(func, c, b, eps)
a = float(input())
b = float(input())
eps = float(input())
print(Root(f, a, b, eps))
```
